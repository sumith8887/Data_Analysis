# Matplotlib - Python visualization library

## 2. Import Matplotlib 

    "import matplotlib.pyplot as plt"

```python
import matplotlib.pyplot as plt 
```

## 3. Line Plot Example

```python
x = [1,2,3,4,5]

y = [10,20,25,30,40]



plt.plot(x, y)

plt.title("Line Plot Example")

plt.xlabel("X-axis")

plt.ylabel("Y-axis")

plt.show()
```

## 4. Bar Plot Example

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.bar(categories, values, color='green')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## 5. Scatter Plot Example

```python
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]
 
plt.scatter(x, y, color='darkgreen', marker='*')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## 6. Histogram Example

```python
data = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
 
plt.hist(data, bins=5, color='lightblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

## 7. Customized Plot Example

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

## 8. Save Plot Example

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
plt.plot(x, y)
plt.title("Save Plot Example")
plt.savefig("my_plot.png")  # Saves in current directory
plt.show()
```

## 9. Pie Chart Example

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

## 10. Box Plot Example

```python
data = [25, 30, 35, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example (Statistical Distribution)")
plt.ylabel("Values")
plt.show()
```

## 11. Violin Plot Example

```python
import numpy as np
data = np.random.normal(0, 1, 100)
plt.violinplot(data)
plt.title("Violin Plot Example")
plt.show()
```

## 12. Stem Plot Example

```python
x = np.linspace(0, 2*np.pi, 10)
y = np.cos(x)
plt.stem(x, y)
plt.title("Stem Plot Example")
plt.show()
```

## 13. Step Plot Example

```python
x = [1,2,3,4,5]
y = [2,3,5,6,8]
plt.step(x, y, where='mid', color='magenta')
plt.title("Step Plot Example")
plt.show()
```

## 14. Area Plot Example

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.fill_between(x, y, color='lightblue', alpha=0.6)
plt.title("Area Plot Example")
plt.show()
```

## 15. Contour Plot Example

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

## 16. Polar Plot Example

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example\n\n")
plt.show()
```

## 17. Error Bar Plot Example

```python
x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

## 18. Horizontal Bar Plot

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.show()
```

## 19. Stack Plot Example

```python
days = [1,2,3,4,5]
eating = [2,3,4,3,4]
sleeping = [8,7,8,6,7]
working = [8,9,7,8,8]
 
plt.stackplot(days, eating, sleeping, working, labels=['Eating','Sleeping','Working'])
plt.legend()
plt.title("Stack Plot Example")
plt.show()
```

## 20. Hexbin Plot Example

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='Blues')
plt.colorbar()
plt.title("Hexbin Plot Example")
plt.show()
```

## 21. 3D Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
```

## 22. 3D Line Plot

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t
 
ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
 
ax.scatter(x, y, z, color='red', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
 
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='orange')
ax.set_title("3D Sphere Surface")
plt.show()
```

```python
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
```

### Hands-On Exercise
 
1.Create a line plot of student marks across 5 exams.
 
2.Plot a bar chart of sales of 4 products.
 
3.Make a scatter plot of Age vs Salary.
 
4.Draw a histogram of marks scored by students in a class.
 
5.Customize a plot with different colors, markers, and line styles.

```python
exams = ['ML', 'ACD', 'CN', 'WT', 'IP']
marks = [78, 82, 69, 90, 85]

plt.figure(figsize=(6, 4))
plt.plot(exams, marks, marker='o', color='blue', linestyle='-')
plt.title('Student Marks Across 5 Exams')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
products = ['Books', 'Pencils', 'Pens', 'A4 sheets']
sales = [250, 180, 300, 220]

plt.figure(figsize=(6, 4))
plt.bar(products, sales, color=["skyblue"])
plt.title('Sales of 4 Products')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
age = [22, 25, 28, 30, 35, 40, 45, 50]
salary = [25000, 30000, 35000, 40000, 48000, 55000, 60000, 65000]

plt.figure(figsize=(6, 4))
plt.scatter(age, salary, color='green', marker='s')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
marks = [82,87,84,76,66,73,68,83,61,90,81,83,79,81,77]
 
plt.hist(marks, bins=5, color='lightblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 6, 11]
y2 = [1, 2, 4, 7, 9]

plt.figure(figsize=(6, 4))
plt.plot(x, y1, color='purple', marker='v', linestyle='--', label='Line 1')
plt.plot(x, y2, color='darkblue', marker='o', linestyle=':', label='Line 2')
plt.title('Customized Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

exams = ['ML', 'ACD', 'CN', 'WT', 'IP']
marks = [78, 82, 69, 90, 85]
axes[0].plot(exams, marks, color='blue', marker='o')
axes[0].set_title('Line Plot of Marks')
axes[0].set_xlabel('Exams')
axes[0].set_ylabel('Marks')

axes[1].bar(products, sales, color='orange')
axes[1].set_title('Bar Plot of Sales')
axes[1].set_xlabel('Products')
axes[1].set_ylabel('Sales')

plt.tight_layout()
plt.show()
```
