# Seaborn 

## 1. what is seaborn ?

seaborn is python library, that builds on the top of matplotlib library that makes the statistical data easy and beautiful.

## Why we need it ?

1. for creating plots like violin plots, box plots

```python
## 4. Histograms & Countplots
# Histogram
import matplotlib.pyplot as plt 
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
plt.figure(figsize=(8,5))
titanic = sns.load_dataset('titanic')

sns.countplot(data=titanic, x='class', hue='sex')
plt.title('Passenger Distribution by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x='age', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x='class', y='fare', palette='Set2')
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=titanic, x='age', y='fare', hue='class', style='sex')
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Class')
plt.show()
```

```python
plt.figure(figsize=(8,5))
corr = titanic[['survived', 'pclass', 'age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Titanic Dataset)')
plt.show()
```
